```python
input_dir = r"C:\Users\dlsat\Downloads\dogs-vs-cats\train\train"
data =[]
lables=[]
for filename in os.listdir(input_dir):
    img = imread(os.path.join(input_dir, filename))
    if img is not None:
        img = resize(img,(32,32))
        img = rgb2gray(img)
        data.append(img.flatten())
        label = 0 if 'cat' in filename else 1
        lables.append(label)
```

```python
data[1]
```

```python
lables[1]
```

```python
data = np.asarray(data)
lables = np.asarray(lables)
```

```python
data.shape
```

```python
lables
```

```python
data_test = []
test_dir = r"C:\Users\dlsat\Downloads\dogs-vs-cats\test1\test1"
for i in os.listdir(test_dir):
    img = imread(os.path.join(test_dir, i))
    if img is not None:
        img = resize(img,(32,32))
        img = rgb2gray(img)
        data_test.append(img.flatten())
```

```python
data_test = np.asarray(data_test)
```

```python
from sklearn.svm import SVC
svm_model = SVC()
```

```python
svm_model.fit(data,lables)
```

```python
y_pred = svm_model.predict(data_test)
```

```python
y_test = pd.read_csv(r"C:\Users\dlsat\Downloads\dogs-vs-cats\sampleSubmission.csv")
y_test = y_test.iloc[:,1]
y_test = np.asarray(y_test)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
```
